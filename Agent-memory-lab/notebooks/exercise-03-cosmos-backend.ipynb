# Exercise 03 Notebook: Cosmos Backend Validation

- Why: Move from local persistence to cloud persistence in guided steps.
- Outcome: You validate Cosmos-backed memory and cross-run durability.

```python
from pathlib import Path
import subprocess
from dotenv import dotenv_values

def find_repo_root(start: Path) -> Path:
    current = start.resolve()
    for candidate in [current] + list(current.parents):
        if (candidate / 'pyproject.toml').exists() and (candidate / 'demo').exists():
            return candidate
    raise RuntimeError('Could not locate repo root')

REPO_ROOT = find_repo_root(Path.cwd())
print('Repo root:', REPO_ROOT)

def run_cmd(cmd: list[str], cwd: Path = REPO_ROOT):
    print('>', ' '.join(cmd))
    proc = subprocess.run(cmd, cwd=str(cwd), text=True, capture_output=True)
    print(proc.stdout)
    if proc.stderr:
        print(proc.stderr)
    if proc.returncode != 0:
        raise RuntimeError(f'Command failed with exit code {proc.returncode}')
```

## Step 1: Validate required environment values
Why: Cosmos and Azure OpenAI settings are required before run.
Outcome: You know if this environment is ready or blocked.

```python
env_file = REPO_ROOT / '.env'
sample_file = REPO_ROOT / '.env.example'
selected = env_file if env_file.exists() else sample_file
cfg = dotenv_values(selected) if selected.exists() else {}
required = [
    'AZURE_OPENAI_ENDPOINT',
    'AZURE_OPENAI_API_KEY',
    'COSMOS_ENDPOINT',
    'COSMOS_CONNECTION_STRING',
    'AZURE_COSMOS_ENDPOINT',
    'AZURE_COSMOS_CONNECTION_STRING',
]
print('Using env file:', selected)
for key in required:
    print(f'{key}:', 'set' if cfg.get(key) else 'missing')
```

## Step 2: Run Demo 04 (first run)
Why: Establish baseline Cosmos behavior.
Outcome: You capture cloud-backed initialization and recall output.

```python
run_cmd(['uv', 'sync', '--extra', 'dev'])
run_cmd(['uv', 'run', 'python', 'demo/04_cosmosdb.py'])
```

## Step 3: Re-run Demo 04
Why: Proves persisted memory across runs.
Outcome: You can provide cross-run durability evidence.

```python
run_cmd(['uv', 'run', 'python', 'demo/04_cosmosdb.py'])
```
